```python
import numpy as np
import pandas as pd
import matplotlib as plt
import nltk
import re
import string
nltk.download('stopwords')
ps = nltk.PorterStemmer()
es_sword = nltk.corpus.stopwords.words("english")
```

```python
#importation du donner
data = pd.read_csv("datasets/Book.csv")
data
```

```python
#compter la longueur du text
data["longueur_du_text"] = data["summary"].apply(lambda x : len(x) - x.count(" ") )
data.sample(50)
```

```python
#Fonction pour calculer le pourcentage de ponctuation
def nombre_de_ponctuation(text):
    ponctuation = [1 for ponc in text if ponc in string.punctuation]
    resultat = sum(ponctuation)
    taille = len(text) - text.count(" ")
    return (resultat/taille)*100
```

```python
data["pourcentage_ponctuation"] = data["summary"].apply(lambda x: nombre_de_ponctuation(x))
data.sample(50)
```

```python
#foction pour nettoyer le donner
def clean_text(text):
    ponctuation = "".join([ponc for ponc in text if ponc not in string.punctuation])
    tolken = re.split("\W+",ponctuation)
    resultat = [ps.stem(mot) for mot in tolken if mot not in es_sword]
    return resultat
```

```python
data["new_summary"] = data["summary"].apply(lambda x: clean_text(x))
data.sample(50)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import f1_score,confusion_matrix,classification_report
```

```python
""" c'est la qu'on vectorise le caratere summary afin d'obtenir les chiffre"""
vecteur_full = CountVectorizer(analyzer=clean_text)
vecteur_final = vecteur_full.fit_transform(data["summary"])
print(vecteur_final.shape)
```

```python
df = pd.concat([pd.DataFrame(vecteur_final.toarray()),data["longueur_du_text"],data["pourcentage_ponctuation"]],axis=1)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import precision_recall_fscore_support as score
```

```python
#donner entrainement et test
_train,x_test,y_train,y_test = train_test_split(df,data["genre"],test_size=0.2,random_state=0)
```

```python
#model du ML
model = SVC(kernel="linear")
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
""" la obtiner notre resultat qui predit que tous les caraterique peuve t'etre predir avec plus de 50% d'exactitude les categorie de livre"""
y_predict = model.predict(x_test)
print(confusion_matrix(y_test,y_predict))
print(classification_report(y_test,y_predict))
```
